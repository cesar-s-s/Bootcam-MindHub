游닗 Crear una nueva columna llamada ``situacion`` en funcion del indice del pais

| Puntaje | Situaci칩n      |
|---------|----------------|
| 0-15    | Buena          |
| 15-30   | Mas bien buena |
| 30-45   | Problematica   |
| 45-60   | Dificil        |
| 60-100  | Muy grave      |


```python
import numpy as np
final["situacion"] = pd.cut(final['indice'],
                 [0, 15, 30, 45, 60, np.inf],
                 labels=['buena','mas bien buena','problematica', 'dificil', 'muy grave']
          )
```

游닗 Que region tiene la mayor cantidad de situaciones buenas o mas bien buenas?

```python
final.loc[final['situacion'].isin(['buena','mas bien buena'])].groupby("REGION")["REGION"].count().sort_values(ascending=False)
```
