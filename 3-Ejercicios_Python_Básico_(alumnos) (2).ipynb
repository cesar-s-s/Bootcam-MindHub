### 2.2

Escribir un programa que pida al usuario una palabra y muestre por pantalla el número de veces que contiene cada vocal.


```python
palabra = str(input('Hola, ingrese la palabra: '))

vocal_a = 0
vocal_b = 0
vocal_c = 0
vocal_d = 0
vocal_e = 0

for i in palabra:
  vocal = i

  if vocal =='a':
    vocal_a  +=1

  elif vocal == 'b':
    vocal_b +=1

  elif vocal == 'c':
    vocal_c +=1

  elif vocal == 'd':
    vocal_d +=1

  elif vocal == 'e':
    vocal_e +=1


print(f'La palabra: "{palabra}" contiene la siguiente cantidad de vocales a: {vocal_a}, vocales b: {vocal_b}, vocales c: {vocal_c}, vocales d: {vocal_d}, vocales e: {vocal_e},')
```

### 2.7
Escribir una funcion que tome un string como parametro y devuelva la palabra mas larga en dicho string


```python
oracion = "Cuando era niño leí el libro del Corsario Negro, me encantó"
```

```python
def palabra_larga(x):
    palabras = x.split()

    palabra_mas_larga = palabras[0]

    for i in palabras[1:]:
        if len(i) > len(palabra_mas_larga):
            palabra_mas_larga = i

    return palabra_mas_larga

oracion = "Cuando era niño leí el libro del Corsario Negro, me encantó"
palabra_ganadora = palabra_larga(oracion)

print(f'La palabra ganadora es: {palabra_ganadora}')
```

### 2.9
Escribir una funcion que dado una lista de numeros devuelva la multiplicacion de todos ellos


```python
def multiplicar (x):

  resultado = 1

  for i in x:
    resultado *= i
  return resultado

lista = [2, 4, 6, 8]

resultado_total = multiplicar(lista)

print(f'El resultado de la multiplicación es: {resultado_total}')
```
